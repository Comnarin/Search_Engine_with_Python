```python
import logging
from urllib.parse import urljoin
import requests
from bs4 import BeautifulSoup

logging.basicConfig(
    format='%(asctime)s %(levelname)s:%(message)s',
    level=logging.INFO)

class Crawler:

    def __init__(self, urls=[]):
        self.visited_urls = []
        self.urls_to_visit = urls

    def download_url(self, url):
        return requests.get(url).text

    def get_linked_urls(self, url, html):
        soup = BeautifulSoup(html, 'html.parser')
        for link in soup.find_all('a'):
            path = link.get('href')
            if path and path.startswith('/'):
                path = urljoin(url, path)
            yield path

    def add_url_to_visit(self, url):
        if url not in self.visited_urls and url not in self.urls_to_visit:
            self.urls_to_visit.append(url)

    def crawl(self, url):
        html = self.download_url(url)
        for url in self.get_linked_urls(url, html):
            self.add_url_to_visit(url)

    def run(self):
        while self.urls_to_visit:
            url = self.urls_to_visit.pop(0)
            logging.info(f'Crawling: {url}')
            try:
                self.crawl(url)
            except Exception:
                logging.exception(f'Failed to crawl: {url}')
            finally:
                self.visited_urls.append(url)


Crawler(urls=['https://www.bbc.com']).run()
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse

class LinkExtractor:
    def __init__(self, url):
        self.url = url
        self.links_list = []

    def extract_links(self):
        try:
            data = requests.get(self.url)
            data.encoding ='utf-8'
            soup = BeautifulSoup(data.text, 'html.parser')
            x = soup.find_all("a",href=True)
            for i in x:
                parsed_url = urlparse(i.get('href'))
                if parsed_url.scheme and parsed_url.netloc:
                    self.links_list.append(i.get('href'))
            return self.links_list
        except Exception as e:
            print(f'An error occurred: {e}')
            return []

extractor = LinkExtractor("https://www.bbc.com")
links = extractor.extract_links()
print(links)
```

```python
import requests
import sqlite3
from bs4 import BeautifulSoup
from urllib.parse import urlparse

# Create a connection to the database
conn = sqlite3.connect('crawler.db')

# Create a table to store the data
conn.execute('''CREATE TABLE IF NOT EXISTS links (url text, depth int)''')

class LinkExtractor:
    def __init__(self, url, depth):
        self.url = url
        self.depth = depth
        self.links_list = []

    def extract_links(self):
        try:
            # Make an HTTP GET request to the URL
            data = requests.get(self.url)
            data.encoding ='utf-8'
            # Parse the HTML response
            soup = BeautifulSoup(data.text, 'html.parser')
            # Extract the links from the 'a' elements with a 'href' attribute
            x = soup.find_all("a",href=True)
            for i in x:
                parsed_url = urlparse(i.get('href'))
                if parsed_url.scheme and parsed_url.netloc:
                    self.links_list.append(i.get('href'))
            # Insert the URL and depth into the database
            conn.execute("INSERT INTO links (url, depth) VALUES (?, ?)", (self.url, self.depth))
            conn.commit()
            return self.links_list
        except Exception as e:
            print(f'An error occurred: {e}')
            return []

def crawl(start_url, max_depth):
    extractor = LinkExtractor(start_url, 0)
    links = extractor.extract_links()
    for link in links:
        if extractor.depth < max_depth:
            extractor = LinkExtractor(link, extractor.depth + 1)
            sublinks = extractor.extract_links()
            links.extend(sublinks)
    # Remove duplicates
    unique_links = list(set(links))
    return unique_links


start_url = "https://www.bbc.com"
max_depth = 3
all_links = crawl(start_url, max_depth)
print(all_links)

# Close the connection to the database
conn.close()
```
